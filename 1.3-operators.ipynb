```python
## Arithmetic Operations

a = 10
b = 5
add_result = a+b   #Addition
print(add_result)

sub_result = a-b   #Subtraction
print(sub_result)

mult_result = a*b  #Multiplication
print(mult_result)

div_result = a/b  #Division
print(div_result)

floor_div_result = a//b # Floor division
print(floor_div_result)

modulus_result = a%b # Modulus operation gives remainder
print(modulus_result)

exponent_result = a**b # Exponentiation operation (a to the power b) 
print(exponent_result)
```

#### Comparison Operators

```python
# Comparison Operators


## Equal to ==
a = 10
b = 10
print(a==b)

str1 = "Ravi"
str2 = "Ravi"
print(str1==str2)


str1 = "Ravi"
str2 = "ravi"
print(str1==str2)

```

```python
## Not equal to !=
str1 = "Hello"
str2 = "Hi"
print(str1!=str2)

str3 = "Ramesh"
str4 = "Ramesh"
print(str3!=str4)
```

```python
# Greater than >
num1 = 45
num2 = 50
print(num1 > num2)
```

```python
# Less than <
num3 = 10
num4 = 15
print(num3<num4)
```

```python
## Greater than or equal to >=
num1 = 45
num2 = 45
print(num1 >= num2)
```

```python
## Lesser than or equal to >=
num1 = 45
num2 = 45
print(num1 <= num2)
```

#### Logical Operators

```python
## Logical operators- and, or, not

# and
X=True
Y=True
print(X and Y)


X=True
Y=False
print(X and Y)
```

```python
X=False
Y=True
print(X and Y)


X=False
Y=False
print(X and Y)
```

```python
# or
X=True
Y=True
print(X or Y)


X=True
Y=False
print(X or Y)
```

```python
X=False
Y=True
print(X or Y)


X=False
Y=False
print(X or Y)
```

```python
# Not
x = True
print(not x)

x=False
print(not x)
```

```python
# Simple Calculator

num1 = float(input("Enter first number : "))
num2 = float(input("Enter second number : "))

addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2
exponentiation = num1 ** num2

print("Addition :",addition)
print("Subtraction :",subtraction)
print("Multiplication :",multiplication)
print("Division :",division)
print("Floor Division :",floor_division)
print("Modulus :",modulus)
print("Exponentiation :",exponentiation)
```
