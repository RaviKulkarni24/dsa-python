#### Variables
Variables are fundamental elements in programming used to store data that can be referenced and manipulated in a program. In python, variables are created when you assign a value to them, and they do not need explicit declaration to reserve memory space. The declaration happens automatically when you assign a value to a variable.

```python
## Declaring and assigning variables
age = 18
height = 5.9
name = "Joy"
isStudent = True

## printing the variables
print("Age :",age)
print("Height :",height)
print("Name :",name)
print("Is student? :",isStudent)
```

```python
## Naming conventions

## Variable names should be descriptive
## Variables must start with a letter or an underscore '_' and followed by letters,numbers and underscore
## Variable names are case sensitive.


## valid variable names
first_name = "Ravi"
last_name = "Kulkarni"



##invalid Variable names
## 2age = 32 
## first-name = "Hi"
## @name = "K"

```

```python
## Case sensitivity
name = "Ravi"
Name = "Kulkarni"
# name and Name are different.
```

```python
## Python is dynamically typed, meaning type of variable is determined at runtime.

# int
age = 20
print(type(age))

# float
height = 5.9
print(type(height))

# string
name = "Ravi"
print(type(name))

## Bool
is_Student = True
print(type(is_Student))
```

```python
## Type checking and conversion
age = 23
print(type(age))

# Age is integer variable and it is converted to string
age_str = str(age)
print(age_str)
print(type(age_str))
```

```python
## Possible type conversions


age = '20'
print(type(age))
ageInt = int(age)
print(type(ageInt))
```

```python
## float can be converted to int and vice versa.
height = 5.11
print(type(height))
heightInt = int(height)
print(heightInt)
print(type(heightInt))
```

```python
## Dynamic Typing
## Python allows the type of a variable to change as the program executes.
var = 10 #int
print(var,type(var))

var = "HELLO"
print(var,type(var))

var = 3.14
print(var,type(var))
```
